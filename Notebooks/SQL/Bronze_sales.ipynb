```python
# 1. Read the CSV customer_info into a PySpark DataFrame
df_customer_data = spark.read.format("csv") \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .load("/Volumes/sports_sales/00_bronze/bronze_layer/cust_info.csv")

# 2. Write the DataFrame as a Delta table
df_customer_data.write.mode("overwrite").saveAsTable("sports_sales.00_bronze.crm_cust_info")

df_customer_data.head()
```
